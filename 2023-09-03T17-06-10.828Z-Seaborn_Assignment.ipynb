**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
sns.boxplot(x="day", y="tip", data=tips,hue = 'sex', palette={"Male": "blue", "Female": "green"})
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.**

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
sns.distplot(titanic['fare'], kde=False,color='green')
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
sns.boxplot(x="class", y="age", data=titanic,hue = 'alive', palette={"no": "blue", "yes": "magenta"})
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

**Recreate the plot below. Use titanic dataset. ** 

```python
sns.boxplot(x="class", y="age", data=titanic,palette=["blue", "skyblue", "yellow"])#,palette='rainbow')
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
sns.countplot(x="sex", hue="class", data=titanic, palette={"First": "blue", "Second": "green","Third":"red"})

# Add labels and a title
plt.xlabel("sex")
plt.ylabel("Count")

# Show the plot
plt.show()
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
titanic = sns.load_dataset("titanic")

numeric_columns = titanic.select_dtypes(include=['number'])

correlation = numeric_columns.corr()
sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt=".2f")

# Show the plot
plt.show()
```
